```python
from random import choice as choice

from IPython.display import clear_output
```

```python
def choose_word():
    '''Загружает файл с темой и выбирает слово.'''
    while True:
        list_name = input('''Выберите тему.
        Введите 1, чтобы выбрать "Аспект".
        Введите 2, чтобы выбрать "Актантные деривации".
        Введите 3, чтобы выбрать "Слова, которые притворяются лингвистическими терминами".
        ''')

        try:
            with open(list_name + '.txt', encoding='utf-8') as f:
                words = f.read().split()
        except FileNotFoundError:
            print('Введите 1, 2 или 3! Попробуем ещё раз.')
            continue

        return choice(words)

def check_word(word, gamer_letters):
    '''Создаёт строку с отгаданными и скрытыми буквами.'''
    
    letters = list(word)
    covered_word = ''
    
    for letter in letters:
        if letter in gamer_letters:
            covered_word = covered_word + letter + ' '
        else:
            covered_word += '_ '
    
    return covered_word

def print_tries(tries):
    '''Выводит количество попыток с верным управлением.'''
    
    if tries == 1:
        print('Осталась 1 попытка.')
    elif tries == 0 or tries > 4:
        print('Осталось %d попыток.' % (tries))
    else:
        print('Осталось %d попытки.' % (tries))  

def suggest_more():
    '''Предлагает сыграть ещё.'''
    answer = input('''Хотите сыграть ещё?
    Если да, нажмите Enter.
    Если нет — введите что угодно и нажмите Enter.\n''')
    if answer == '':
        start_game()
        
def draw_person(tries):
    if tries == 6:
        print(''' |\n''')
    if tries == 5:
        print(' |\n o\n')
    if tries == 4:
        print(' |\n o\n/\n')
    if tries == 3:
        print(' |\n o\n/|\n')
    if tries == 2:
        print(' |\n o\n/|\\\n')
    if tries == 1:
        print(' |\n o\n/|\\\n/\n')
    if tries == 0:
        print(' |\n o\n/|\\\n/ \\\n')

def ask_and_check(word):
    '''Спрашивает буквы, проверяет, подходят ли они.'''
    russian_letters = list('абвгдеёжзийклмнопрстуфхцчшщъыьэюя')
    tries = 7
    gamer_letters = []
    
    print('У вас есть 7 попыток, чтобы отгадать слово из %d букв.\n' % (len(word)))
    
    while True:
        print(check_word(word, gamer_letters))
        
        if check_word(word, gamer_letters).replace(' ', '') == word:
            print('Ура, вы победили!')
            break
             
        print_tries(tries)
        
        if tries == 0:
            print('Увы, вы проиграли(\n')
            print('Это было слово "%s".\n' % (word))
            break  
        
        gamer_letter = input('Введите кириллическую букву: ').lower()
        
        # clear_output(True)

        if gamer_letter == '':
            print('Нужно ввести хоть что-то...\n')
        elif gamer_letter in gamer_letters:
            print('Эта буква уже была, выберите другую.\n')
        elif gamer_letter not in russian_letters:
            print('Вы ввели не русскую букву, а что-то иное.\n')
        elif gamer_letter not in word:
            print('Увы, в слове этой буквы нет(\n')
            gamer_letters += gamer_letter
            tries -= 1
            draw_person(tries)
        else:
            print('Ура, вы угадали букву!\n')
            gamer_letters += gamer_letter
    
    suggest_more()
    
def start_game():
    '''Запускает игру.'''
    ask_and_check(choose_word())
```

```python
start_game()
```
